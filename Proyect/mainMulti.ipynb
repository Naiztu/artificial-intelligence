## Data preprocessing

```python

df = df.drop(['filename'], axis=1)
df['label'] = df['label'].replace({'blues':0, 'classical':1, 'country':2, 'disco':3, 'hiphop':4, 'jazz':5, 'metal':6, 'pop':7, 'reggae':8, 'rock':9})
df
```

## Apply escalation techniques.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_df = pd.DataFrame(scaler.fit_transform(df.drop('label', axis=1)), columns=['tempo', 'beats', 'chroma_stft', 'rmse', 'spectral_centroid',
       'spectral_bandwidth', 'rolloff', 'zero_crossing_rate', 'mfcc1', 'mfcc2',
       'mfcc3', 'mfcc4', 'mfcc5', 'mfcc6', 'mfcc7', 'mfcc8', 'mfcc9', 'mfcc10',
       'mfcc11', 'mfcc12', 'mfcc13', 'mfcc14', 'mfcc15', 'mfcc16', 'mfcc17',
       'mfcc18', 'mfcc19', 'mfcc20'])
```

```python
scaled_df['label'] = df['label']
scaled_df
```

```python
from sklearn.model_selection import train_test_split

train_df_x, test_df_x, train_df_y, test_df_y  = train_test_split(scaled_df.drop('label', axis=1), scaled_df['label'], test_size=0.2, stratify=scaled_df['label'])
train_df_x, val_df_x, train_df_y, val_df_y = train_test_split(train_df_x, train_df_y, test_size=0.1, stratify=train_df_y)
```

```python
train_df_x
```

```python
model_name = 'model-multi'
best_accuracy = 0
num_classes = 10
```

```python
import tensorflow as tf
from tensorflow.keras.utils import to_categorical

model = tf.keras.models.load_model(model_name +'.h5')
best_accuracy = model.evaluate(test_df_x, to_categorical(test_df_y, num_classes=num_classes))[1] if model else 0
```

```python
import optuna
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

def objective(trial):
    global best_accuracy

    units_lstm = trial.suggest_int('units_lstm', 32, 125)
    learning_rate = trial.suggest_float('learning_rate', 1e-5, 1e-2)

    model = Sequential()

    model.add(LSTM(units_lstm, input_shape=(train_df_x.shape[1], 1), activation='tanh', return_sequences=True))
    
    layers_number_lstm = trial.suggest_int('layers_number_lstm', 0, 1)
    for _ in range(layers_number_lstm):
        model.add(LSTM(units_lstm, activation='tanh', return_sequences=True))
    model.add(LSTM(units_lstm, activation='tanh'))

    layers_number_dense = trial.suggest_int('layers_number_dense', 1, 2)
    units = trial.suggest_int('units', 32, 512)
    
    for _ in range(layers_number_dense):
        model.add(Dense(units, activation='relu'))

    model.add(Dense(num_classes, activation='relu'))

    model.compile(optimizer=Adam(learning_rate=learning_rate), loss='categorical_crossentropy', metrics=['accuracy'])

    history = model.fit(train_df_x, to_categorical(train_df_y, num_classes=num_classes), epochs=50, validation_data=(val_df_x,  to_categorical(val_df_y, num_classes=num_classes)), verbose=0)

    acuuracy = history.history['val_accuracy'][-1]
    if  acuuracy > best_accuracy:
        model.save(model_name+'.h5')
        f = open('best_model.txt', 'w')
        f.write('accuracy : '+ str(history.history['accuracy'][-1]))
        f.write('\n')
        f.write('loss : '+ str(history.history['loss'][-1]))
        f.write('\n')
        f.write('val_accuracy : '+ str(history.history['val_accuracy'][-1]))
        f.write('\n')
        f.write('val_loss : '+ str(history.history['val_loss'][-1]))
        f.write('\n\nHIPERPARAMETERS\n\n')
        f.write('units_lstm : '+ str(units_lstm))
        f.write('\n')
        f.write('learning_rate : '+ str(learning_rate))
        f.write('\n')
        f.write('layers_number_lstm : '+ str(layers_number_lstm))
        f.write('\n')
        f.write('layers_number_dense : '+ str(layers_number_dense))
        f.write('\n')
        f.write('units : '+ str(units))
        f.close()
        plt.plot(history.history['accuracy'])
        plt.plot(history.history['val_accuracy'])
        plt.title('model accuracy')
        plt.ylabel('accuracy')
        plt.xlabel('epoch')
        plt.legend(['train', 'val'], loc='upper left')
        plt.savefig('accuracy-'+model_name+'.png')
        plt.close()
        plt.plot(history.history['loss'])
        plt.plot(history.history['val_loss'])
        plt.title('model loss')
        plt.ylabel('loss')
        plt.xlabel('epoch')
        plt.legend(['train', 'val'], loc='upper left')
        plt.savefig('loss-'+model_name+'.png')
        plt.close()

        best_accuracy = acuuracy


    # Return the validation accuracy
    return history.history['val_accuracy'][-1]
```

```python
study = optuna.create_study(direction='maximize')

train_flag = True

if train_flag:
    study.optimize(objective, n_trials=50, n_jobs=-1)
```

```python

import numpy as np
predicted_labels = np.argmax(predicted, axis=1)
predicted_labels
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

import matplotlib.pyplot as plt

cm = confusion_matrix(test_df_y, predicted_labels)

plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('True labels')
plt.ylabel('Predicted labels')
plt.show()
```

```python
labels = np.unique(df['label'])

for label in labels:
    true = test_df_y == label
    predicted = predicted_labels == label

    tn, fp, fn, tp = confusion_matrix(true, predicted).ravel()

    tpr = tp / (tp + fn)
    fpr = fp / (fp + tn)

    print(f'Label {label}:')
    print(f'True positive rate: {tpr}')
    print(f'False positive rate: {fpr}')
```
